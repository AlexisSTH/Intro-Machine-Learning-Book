```python
import mglearn
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# On crée des données fictives qui ressemblent à une vague
X, y = mglearn.datasets.make_wave(n_samples=60)

# 1. On divise (Split)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# 2. On choisit le modèle
# LinearRegression, méthode classique (Moindres Carrés Ordinaires)
lr = LinearRegression()

# 3. On entraîne (Fit) Le modèle cherche la meilleure ligne droite
lr.fit(X_train, y_train)

# 4. On regarde ce qu'il a trouvé (Les paramètres de la ligne)
print("Pente (w) calculée :", lr.coef_)
print("Point de départ (b) calculé :", lr.intercept_)

# 5. On regarde les scores (R²)
print("\n--- Résultats ---")
print("Score sur l'entraînement (Blue line) : {:.2f}".format(lr.score(X_train, y_train)))
print("Score sur le test (Orange line)      : {:.2f}".format(lr.score(X_test, y_test)))
print("underfitting : modèle trop simple")
#Visualisation
plt.plot(X_train, y_train, 'o', label="Training data") # Les points
plt.plot(X_test, y_test, 'v', label="Test data")       # Les points tests
plt.plot(X, lr.predict(X), label="Prediction Line")    # LA LIGNE
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import mglearn

# On charge le jeu de données "Extended Boston"
X, y = mglearn.datasets.load_extended_boston()

# On sépare (toujours avec random_state=0 pour avoir les mêmes résultats)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# On entraîne le modèle linéaire classique
lr = LinearRegression().fit(X_train, y_train)

# On affiche les scores
print("Score entraînement: {:.2f}".format(lr.score(X_train, y_train)))
print("Score test: {:.2f}".format(lr.score(X_test, y_test)))
print("Overfitting : score non équilibré (Sur-apprentissage)")
```

```python
from sklearn.linear_model import Ridge

# On entraîne le modèle Ridge
ridge = Ridge().fit(X_train, y_train)

# On affiche les scores
print("Score entraînement (Ridge): {:.2f}".format(ridge.score(X_train, y_train)))
print("Score test (Ridge)        : {:.2f}".format(ridge.score(X_test, y_test)))
print("Score équilibré")
```

```python
# On augmente le contrainte (alpha = 10)
ridge10 = Ridge(alpha=10).fit(X_train, y_train)

print("Score entraînement (alpha=10): {:.2f}".format(ridge10.score(X_train, y_train)))
print("Score test (alpha=10)        : {:.2f}".format(ridge10.score(X_test, y_test)))
```

```python
# Ridge avec alpha=0.1 (Très peu de contraintes)
ridge01 = Ridge(alpha=0.1)
ridge01.fit(X_train, y_train)

print(f"Score entraînement (alpha=0.1): {ridge01.score(X_train, y_train):.2f}")
print(f"Score test (alpha=0.1)        : {ridge01.score(X_test, y_test):.2f}")
```
