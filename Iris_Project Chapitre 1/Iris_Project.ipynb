```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

print("Bibliothèques importées avec succès !")
```

```python
# On charge le dataset
iris_dataset = load_iris()

# On affiche ce qu'il contient
print("Noms des cibles (les espèces) :", iris_dataset['target_names'])
print("Noms des caractéristiques (les mesures) :", iris_dataset['feature_names'])
print("Taille des données :", iris_dataset['data'].shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)

print("Taille entrainement :", X_train.shape)
print("Taille test :", X_test.shape)
```

```python
# On choisit le modèle (voisin le plus proche = 1)
knn = KNeighborsClassifier(n_neighbors=1)

# On l'entraîne (La méthode .fit signifie "APPRENDS")
knn.fit(X_train, y_train)

print("Modèle entraîné !")
```

```python
precision = knn.score(X_test, y_test)

print("Précision du modèle : {:.2f}".format(precision))
```

```python
# Nouvelles mesures
nouvelle_fleur = np.array([[5, 2.9, 1, 0.2]])

# Prédiction
prediction = knn.predict(nouvelle_fleur)

print("Espèce prédite :", iris_dataset['target_names'][prediction])
```
